# Data Cleaning

```python
# Rename Columns
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head(10)
```

```python
# Check Missing Values
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values[missing_values > 0]
```

```python
# Handle Missing Values
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Convert 'age' to integer (round then convert)
df['age'] = df['age'].round().astype(int)
df['age'].dtype
```

```python
# Check Duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Illogical Values Detection
negative_values = (df[numeric_cols] < 0).sum()
negative_values[negative_values > 0]
```

```python
# Check details of illogical values
df.loc[df['total_purchases'] < 0, 'total_purchases'].describe()
```

```python
# Change illogical values to NaN
df.loc[df['total_purchases'] < 0, 'total_purchases'] = np.nan
```

```python
# Impute the NaN values with median
df['total_purchases'] = df['total_purchases'].fillna(df['total_purchases'].median())
```

```python
df[df['total_purchases'] < 0]
```

```python
# Illogical Values in Age
df['age'].describe()
```

```python
df['age'] = df['age'].clip(lower=18, upper=90)
```

```python
# Outlier Check
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
```

```python
outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | 
            (df[numeric_cols] > (Q3 + 1.5 * IQR))).sum()

outliers.sort_values(ascending=False)
```

```python
df_clean = df.copy()
```

# Segmenting Customers

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Select features for clustering
features = df_clean[['total_purchases', 'session_duration_avg', 'age']]
```

```python
# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df_clean['cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Visualize the customer segments
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_purchases', y='session_duration_avg', hue='cluster', palette='Set1', data=df_clean)
plt.title('Customer Segmentation Based on Purchases and Session Duration')
plt.show()
```

```python
# Churn rate per cluster
df_clean.groupby('cluster')['churned'].mean()
```

### Churn Rate by Customer Segment

After segmenting customers based on their purchasing behavior and session duration, the churn rate was calculated for each cluster.

The results show clear differences across segments. Cluster 2 has the highest churn rate (around 41%), indicating that customers in this segment are at a much higher risk of churn. On the other hand, Cluster 0 shows the lowest churn rate (around 18%), suggesting a group of more loyal and engaged customers.

This confirms that customer engagement and purchase behavior are strongly related to churn, and that retention strategies should be tailored based on customer segment.

```python
churn_by_cluster = df_clean.groupby('cluster')['churned'].mean()

ax = churn_by_cluster.plot(kind='bar', figsize=(6,4))
ax.set_title('Churn Rate by Cluster')
ax.set_xlabel('Cluster')
ax.set_ylabel('Churn Rate')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)  # biar lurus
plt.tight_layout()
plt.show()
```
