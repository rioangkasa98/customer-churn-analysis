## Data Quality Checks

```python
# Rename Columns
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head(10)
```

```python
# Check Missing Values
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values[missing_values > 0]
```

```python
# Handle Missing Values
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Convert 'age' to integer (round then convert)
df['age'] = df['age'].round().astype(int)
df['age'].dtype
```

```python
# Check Duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Illogical Values Detection
negative_values = (df[numeric_cols] < 0).sum()
negative_values[negative_values > 0]
```

```python
# Check details of illogical values
df.loc[df['total_purchases'] < 0, 'total_purchases'].describe()
```

```python
# Change illogical values to NaN
df.loc[df['total_purchases'] < 0, 'total_purchases'] = np.nan
```

```python
# Impute the NaN values with median
df['total_purchases'] = df['total_purchases'].fillna(df['total_purchases'].median())
```

```python
df[df['total_purchases'] < 0]
```

```python
# Illogical Values in Age
df['age'].describe()
```

```python
df['age'] = df['age'].clip(lower=18, upper=90)
```

```python
# Outlier Check
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
```

```python
outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | 
            (df[numeric_cols] > (Q3 + 1.5 * IQR))).sum()

outliers.sort_values(ascending=False)
```
